```python
class Node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
def insert(root,key):
    if root is None:
        print(f'create root node {key}')
        return Node(key)
    else:
        if key==root.data:
            print('the root node is equal to key')
            return root
        elif key<root.data:
            print(f'{key}<{root.data}--> move left')
            root.left=insert(root.left,key)
        else:
            print(f'{key}>{root.data}--> move right')
            root.right=insert(root.right,key)
    return root
def show_tree(root, space=0, side="Root:"):
    if root:
        print(" " * space + side + str(root.data)) # object reference
        show_tree(root.left, space + 4, "L---")
        show_tree(root.right, space + 4, "R---")
root=None
values=[4,2,6,1,3,5,7]
for key in values:
    root=insert(root,key)
    show_tree(root)
    print('-'*30)
```

```python
def search(root,find_values):
    if root is None:
        print('Node is empty')
        return False
    print(f'check: {root.data}')
    if find_values==root.data:
        print('Found')
        return True
    elif find_values<root.data:   
        return search(root.left,find_values)
    return search (root.right,find_values)
search(root,1)
```
